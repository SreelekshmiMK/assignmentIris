# Visualization

# 1.Histogram

```python
plt.figure()
plt.hist(dataset['SL'],rwidth=0.9)
plt.title('Distribution of SL',fontsize=16)
plt.xticks()
plt.yticks()
```

```python
dataset['PL'].hist()
plt.title('Distribution of PL')
```

```python
dataset['SW'].hist()
plt.title('Distribution of SW')
```

```python
dataset['PW'].hist()
plt.title('Distribution of PW')
```

# 2.Scatter Plot

```python
plt.figure()
plt.scatter(dataset['SL'],dataset['SW'],s=10,c='red')
plt.title('Distribution of SL vs SW',fontsize=16)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.xticks()
plt.yticks()
```

```python
plt.figure()
plt.scatter(dataset['PL'],dataset['PW'],s=10,c='green')
plt.title('Distribution of SL',fontsize=16)
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.xticks()
plt.yticks()
```

# 3.Pair Plot

```python
sns.pairplot(dataset)
```

# 4.Box Plot

```python
sns.boxplot(x='SL',y='SW',data=dataset)
```

# 5.Strip Plot

```python
sns.stripplot(x='PL',y='PW',data=dataset)
```
